# **1.Factor Finder**

```python
# the program can handle expressions
# input by a user, write a program that will ask the user to input an expression, calculate its factors, and print them. Your program should be able to
# handle invalid input by making use of exception handling

from sympy import factor
from sympy.parsing.mathematica import parse_mathematica

while True:
  expression_str = input("Enter an algebraic expression to factor (or 'quit' to exit): ")
  if expression_str.lower() == 'quit':
    break
  try:
    # Attempt to parse the expression. SymPy's parse_mathematica can be
    # robust, but you might need other parsers depending on expected input format.
    expression = parse_mathematica(expression_str)
    factored_expression = factor(expression)
    print(f"The factors of {expression_str} are: {factored_expression}")
  except Exception as e:
    print(f"Invalid input: {e}. Please enter a valid algebraic expression.")
```

# **2.Graphical Equation Solver**

```python
# Graphical Equation Solver Earlier, you learned how to write a program that prompts the user to input an expression such as 3x + 2y − 6 and create the corresponding graph.
# Write a program that asks the user for two expressions and then graphs them both,
# as follows:
# >>> expr1 = input('Enter your first expression in terms of x and y: ')
# >>> expr2 = input('Enter your second expression in terms of x and y: ')
# Now, expr1 and expr2 will store the two expressions input by the user.
# You should convert both of these into SymPy objects using the sympify() step
# in a try...except block.
# All you need to do from here is plot these two expressions instead
# of one.
# Once you’ve completed this, enhance your program to print the
# solution—the pair of x and y values that satisfies both equations. This
# will also be the spot where the two lines on the graph intersect. (Hint:
# Refer to how we used the solve() function earlier to find the solution of a
# system of two linear equations

from sympy import Symbol, sympify, solve
from sympy.plotting import plot
from sympy.core.sympify import SympifyError

# Define the symbols
x = Symbol('x')
y = Symbol('y')

while True:
    expr1_str = input('Enter your first expression in terms of x and y (or "quit" to exit): ')
    if expr1_str.lower() == 'quit':
        break

    expr2_str = input('Enter your second expression in terms of x and y: ')

    try:
        # Convert input strings to SymPy expressions
        # We need to create equations by setting the expressions equal to zero
        expr1 = sympify(expr1_str) - y
        expr2 = sympify(expr2_str) - y

        # Plot the expressions
        # We plot expr1 = y and expr2 = y
        p = plot(expr1 + y, expr2 + y, show=False, legend=True) # Add y back for plotting form
        p[0].label = expr1_str
        p[1].label = expr2_str
        p.show()

        # Solve the system of equations
        # We solve for the point where expr1 = expr2
        solution = solve((expr1, expr2), (x, y))

        if solution:
            print(f"The solution (intersection point) is: {solution}")
        else:
            print("No unique solution found for this system of equations.")

    except SympifyError:
        print("Invalid input: Please enter a valid algebraic expression.")
    except Exception as e:
        print(f"An error occurred: {e}")

```
