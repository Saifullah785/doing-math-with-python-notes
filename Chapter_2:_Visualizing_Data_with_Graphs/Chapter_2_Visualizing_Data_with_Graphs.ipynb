**Plotting with Formulas**

```python
# This cell defines a function `grav_force` that calculates the gravitational force between two bodies based on Newton's Law of Universal Gravitation and then calls this function with specific values.
# prompt: write a formula of Newton’s Law of Universal Gravitation

# # **Newton’s Law of Universal Gravitation**
def grav_force(m1, m2, d):
  G = 6.674*(10**-11) # gravitational constant
  return G*(m1*m2)/(d**2)

grav_force(0.1, 0.1, 0.01)
```

```python
'''
The relationship between gravitational force and distance between two bodies

'''
# This cell imports the matplotlib.pyplot library as plt to be used for plotting gravitational force.
import matplotlib.pyplot as plt
```

```python
# This cell defines two functions: `draw_graph` to plot the gravitational force vs. distance, and `generate_F_r` to calculate gravitational force at different distances and then plot the results. The main execution block calls `generate_F_r`.
# Draw the graph
def draw_graph(x, y):
  plt.plot(x, y, marker='o')
  plt.xlabel('Distance in meters')
  plt.ylabel('Gravitational force in Newtons')
  plt.title('Gravitational force vs. distance')
  plt.show()


def generate_F_r():
  # Generate values for r
  r = range(100, 1001, 50)
  # Empty list to store the calculated values fo F
  F = []

  # constant , G

  G = 6.674*(10**-11)

  # Two masses
  m1 = 0.5
  m2 = 1.5

  # Calculate force and add it to the list, F
  for dist in r:
    force = G * (m1 * m2) / (dist**2)
    F.append(force)

  # Draw the graph
  draw_graph(r, F)


if __name__ == '__main__':
  generate_F_r()


#
```

# **Generating Equally Spaced Floating Point Numbers**

```python
'''
Generate equally spaced floating point numbers between two given values
'''
# This cell defines a function `frange` that generates a list of floating-point numbers within a given range with a specified increment and then calls this function with specific arguments.
def frange(start, final, increment):
  numbers = []
  while start < final:
    numbers.append(start)
    start = start + increment
  return numbers

frange(0.5, 1.0, 0.1)
```

# **Drawing the Trajectory**

```python
'''
Draw the trajectory of a body in projectile motion
'''
# This cell defines functions to calculate and plot the trajectory of a projectile given initial velocity and angle. It includes a function `frange` for generating time intervals and prompts the user for input values.
from matplotlib import pyplot as plt
import math



def draw_graph(x, y):
 plt.plot(x, y)
 plt.xlabel('x-coordinate')
 plt.ylabel('y-coordinate')
 plt.title('Projectile motion of a ball')



def frange(start, final, interval):

  numbers = []
  while start < final:
    numbers.append(start)
    start = start + interval
  return numbers



def draw_trajectory(u, theta):
  theta = math.radians(theta)
  g = 9.8

  # time of flight
  t_flight = 2*u*math.sin(theta)/g
  # time intervals
  intervals = frange(0, t_flight, 0.001)

# list of x and y coordinates
  x = []
  y = []

  for t in intervals:
    x.append(u*math.cos(theta)*t)
    y.append(u*math.sin(theta)*t - 0.5*g*t*t)

  draw_graph(x, y)

if __name__ == '__main__':
  try:
    u = float(input('Enter the initial velocity (m/s): '))
    theta = float(input('Enter the angle of projection (degrees): '))
  except ValueError:
    print('You entered an invalid input')

  else:

    draw_trajectory(u, theta)
    plt.show()
```

# **Comparing the Trajectory at Different Initial Velocities**

```python
# This cell compares the trajectory of a projectile at different initial velocities by calling the `draw_trajectory` function with a list of velocities and then adds a legend to the plot.
if __name__ == '__main__':

 # List of three different initial velocities
 u_list = [20, 40, 60]
 theta = 45
 for u in u_list:

  draw_trajectory(u, theta)

 # Add a legend and show the graph
 plt.legend(['20', '40', '60'])
 plt.show()
```
